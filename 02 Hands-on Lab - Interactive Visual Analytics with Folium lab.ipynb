Now, you can take a look at what are the coordinates for each site.

```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#0000ff', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
# Task 2: Mark the success/failed launches for each site on the map

success = spacex_df[spacex_df['class'] == 1][['Launch Site','Lat','Long','class']].groupby(by=['Launch Site','Lat','Long']).count().reset_index()
unsuccess = spacex_df[spacex_df['class'] == 0][['Launch Site','class']].groupby(by='Launch Site').count().reset_index()
result = success.merge(unsuccess, how='inner', on='Launch Site')
result.columns = ['Launch Site','Lat','Long','success','unsuccess']


success_map = folium.Map(location=nasa_coordinate, zoom_start=5)

for row in result.iterrows():
    coordinate = [row[1][1], row[1][2]]
    sitename = row[1][0]
    success = row[1][3]
    unsuccess = row[1][4]
    success_rate = success / (success+unsuccess)
    circle = folium.Circle(coordinate, radius=50, color='#d35400', fill=True).add_child(folium.Popup(sitename))
    marker = folium.map.Marker(
    coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;width: 100px "><b>%s</b></div>' % (sitename + 
                                                                                          '<br/>Success:' + str(success) +
                                                                                          '<br/>Unsuccess: '+ str(unsuccess) + 
                                                                                          '<br/>Success Rate:'+ (str(success_rate))
                                                                                         )
        
        ,
        )
    )
    success_map.add_child(circle)
    success_map.add_child(marker)

success_map

```

# TASK 3: Calculate the distances between a launch site to its proximities

Next, we need to explore and analyze the proximities of launch sites.

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)

```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
launch_site_lat = 28.562302
launch_site_lon = -80.577356
coastline_lat = 28.56715
coastline_lon = -80.56699
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
coordinate = [28.56715, -80.56699]
distance_marker = folium.Marker(
    coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )
site_map.add_child(distance_marker)
site_map
```

*TODO:* Draw a `PolyLine` between a launch site to the selected coastline point


```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
coordinates = [
    [launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]
]
lines=folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)
```

```python
launch_site_lat = 28.562302
launch_site_lon = -80.577356

city_lat = 28.54351
city_lon = -81.39496

railway_lat = 28.54841
railway_lon = -81.38088

hway_lat = 28.49283
hway_lon = -80.57819

distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_hway = calculate_distance(launch_site_lat, launch_site_lon, hway_lat, hway_lon)

distance_marker_city = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )

distance_marker_railway = folium.Marker(
    [railway_lat, railway_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )

distance_marker_hway = folium.Marker(
    [hway_lat, hway_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )

site_map.add_child(distance_marker_city)
site_map.add_child(distance_marker_railway)
site_map.add_child(distance_marker_hway)

line_coordinates_city = [
    [launch_site_lat, launch_site_lon], [city_lat, city_lon]
]
lines_city=folium.PolyLine(locations=line_coordinates_city, weight=1)

line_coordinates_railway = [
    [launch_site_lat, launch_site_lon], [railway_lat, railway_lon]
]
lines_railway=folium.PolyLine(locations=line_coordinates_railway, weight=1)

line_coordinates_hway = [
    [launch_site_lat, launch_site_lon], [hway_lat, hway_lon]
]
lines_hway=folium.PolyLine(locations=line_coordinates_hway, weight=1)

site_map.add_child(lines_city)
site_map.add_child(lines_railway)
site_map.add_child(lines_hway)
site_map
```
